```python
df['Value'] = df['Value'].str.replace('$','')
df['Value'] = df['Value'].str.replace(',','')
# df['Value'] = df['Value'].str.replace('None', np.nan)
# df['Value'].astype(int)
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvec = CountVectorizer()
cvec.fit(df['Question'])
```

```python
words = pd.DataFrame(cvec.transform(np.array(df['Question']),
             columns=cvec.get_feature_names()))

words.transpose().sort_values(0, ascending=False)
```
